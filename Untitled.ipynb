```python
data_columns = set(unsplit_data)
data_columns.remove("price")
data_columns = list(data_columns)
x_data = unsplit_data[data_columns]
x_data.head()
```

```python
os = x_data['os']
```

```python
brand = []
model = []
rows = x_data.shape[0]
for i in range(0,rows):
    model_brand = x_data.loc[i].at["model"]
    split_arr = model_brand.split()
    brand.append(split_arr[0])
    n = len(split_arr)
    text = []
    for i in range(1,n):
        text.append(split_arr[i])
    model.append(" ".join(text))
x_data['brand'] = brand
x_data['model_only'] = model
x_data = x_data.drop(['model'],axis=1)
x_data.rename(columns={'model_only':'model'},inplace=True)
x_data.head()
```

```python
rows = x_data.shape[0]
for i in range(0,rows):
    hi = "hi"
    card = x_data.loc[i].at["card"]
    if(type(card)==type(hi)):
        if(not card.startswith("Memory")):
            x_data.replace(x_data.loc[i].get('card'),0,inplace=True)
    else:
        continue
x_data['os'] = os
x_data.head()
```

```python
x_data['card'].value_counts()
```

```python
card = []
card_size = []
rows = x_data.shape[0]
for i in range(0,rows):
    card_data = x_data.loc[i].at["card"]
    if(card_data=='Memory Card (Hybrid)'  or card_data=='Memory Card Supported'):
        card.append(1)
        card_size.append(np.nan)
        continue
    elif(card_data == 'Memory Card Not Supported' or card_data==0 or card_data==np.nan):
        card.append(0)
        card_size.append(0)
        continue
    else:
        split_data = str(card_data).split(",")
        if(len(split_data)==2):
            card.append(1)
            card_size.append(split_data[1].strip())
        else:
            card.append(0)
            card_size.append(0)
x_data['new_card'] = card
x_data['card_size'] = card_size
```

```python
x_data = x_data.drop(['card'],axis=1)
```

```python
x_data.rename(columns={'new_card':'card'},inplace=True)
x_data.head(50)
```

```python
ram =[]
rom = []
for i in range(0,rows):
    ram_rom = x_data.loc[i].at['ram']
    if(ram_rom.find("MB")==-1 and ram_rom.find("GB") == -1 and ram_rom.find("TB") ==-1):
        ram.append(np.nan)
    else:
        ram.append(ram_rom)
x_data['ram'] = ram
```

```python
x_data.ram.value_counts()
```

```python
ram =[]
rom = []
for i in range(0,rows):
    ram_rom = x_data.loc[i].at['ram']
    try:
        split_data = ram_rom.split(",")
    except:
        ram.append(np.nan)
        rom.append(np.nan)
        continue
    if(len(split_data)!=2):
        rom_split = split_data[0].split()
        ram.append(np.nan)
        if("GB" in rom_split):
            rom.append(rom_split[0])
        elif("TB" in rom_split):
            rom.append(int(rom_split[0])*1000)
        elif("MB" in rom_split):
            rom.append(int(rom_split[0])*0.001)
        else:
            rom.append(np.nan)
    elif(len(split_data)==2):
        ram_split = split_data[0].split()
        if("GB" in ram_split):
            ram.append(ram_split[0])
        elif("TB" in ram_split):
            ram.append(int(ram_split[0])*1000)
        elif("MB" in ram_split):
            ram.append(int(ram_split[0])*0.001)
        else:
            ram.append(np.nan)
        rom_split = split_data[1].split()
        if("GB" in rom_split):
            rom.append(rom_split[0])
        elif("TB" in rom_split):
            rom.append(int(rom_split[0])*1000)
        elif("MB" in rom_split):
            rom.append(int(rom_split[0])*0.001)
        else:
            rom.append(np.nan)
    else:
        ram.append(np.nan)
        rom.append(np.nan)
x_data['rom'] = rom
x_data['ram'] = ram
x_data.head(50)
```
